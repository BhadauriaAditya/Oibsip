```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/Unemployment in India.csv")
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={df.columns[0]:'Region',df.columns[3]:'Unemployment_rate',df.columns[4]:'Employed', df.columns[5]:'labour_participation_rate'})
df.head()
```

```python
df["Region"].unique()
```

```python
df2 = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")

df2
```

```python
df2 = df2.rename(columns={df2.columns[0]:'Region',df2.columns[3]:'Unemployment_rate',df2.columns[4]:'Employed', df2.columns[5]:'labour_participation_rate',})
df2.head()
```

```python
heat_maps = df[['Unemployment_rate','Employed','labour_participation_rate']]

heat_maps = heat_maps.corr()

plt.figure(figsize=(12,7))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True);
```

```python
df2.columns = ["Region", "Date", "Frequency", "Unemployment_rate", "Employed", "Labour_participation_rate", "Area", "Longitude", "Latitude"]

plt.figure(figsize=(14, 10))
plt.title("Unemployment Rate")
sns.histplot(x="Unemployment_rate", hue="Region", data=df)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

region = df2.groupby(["Region"])[['Unemployment_rate', "Employed", "Labour_participation_rate"]].mean()
region = pd.DataFrame(region).reset_index()

fig = px.bar(region, x="Region", y="Employed", title="Average Employed Rate by Region")
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
region = df2.groupby(["Region"])[['Unemployment_rate', "Employed", "Labour_participation_rate"]].mean()
region = pd.DataFrame(region).reset_index()

fig = px.bar(region, x="Region", y="Unemployment_rate", title="Average Unemployed Rate by Region")
fig.update_layout(xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
l = ['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']
s = [50,50,50]
ax.pie(s,labels = l,autopct = '%1.2f%%')
plt.show()
```

```python
unemployment = df2[["Region", "Area", "Unemployment_rate"]]

fig = px.sunburst(unemployment, path=['Area','Region'], values='Unemployment_rate',
                  title= 'Unemployment rate in every State and Region', height=700)
fig.show()
```
